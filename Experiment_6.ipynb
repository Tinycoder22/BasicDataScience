### Plotting 3-D lines & plots.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional line
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# Plot the 3-D line
ax.plot(x, y, z)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

### Scatter Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional scatter plot
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(0, 1, 100)

# Plot the 3-D scatter plot
ax.scatter(x, y, z)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

### Surface plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the surface
ax.plot_surface(X, Y, Z)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

### Density & Contour plots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

# Create a 2-D histogram of the data
plt.hist2d(x, y, bins=30, cmap='Blues')

# Add a color bar
cb = plt.colorbar()
cb.set_label('Counts')

# Add labels
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('2-D Density Plot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

# Create a 2-D histogram of the data
H, xedges, yedges = np.histogram2d(x, y, bins=30)

# Plot the contours
plt.contourf(H.T, extent=[xedges.min(), xedges.max(), yedges.min(), yedges.max()], cmap='Blues')

# Add a color bar
cb = plt.colorbar()
cb.set_label('Counts')

# Add labels
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('2-D Contour Plot')

plt.show()
```

### Changing the camera angle of 3-D plot.

```python
# Generate some data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3-D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool')

# Set the camera angle
ax.view_init(elev=30, azim=45)

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3-D Plot with Camera Angle')

plt.show()
```

### Stacking subplots in one direction.

```python
# Create some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two stacked subplots
fig, axs = plt.subplots(nrows=2, ncols=1)

# Plot the first subplot
axs[0].plot(x, y1, 'r')

# Plot the second subplot
axs[1].plot(x, y2, 'b')

# Add titles and labels
axs[0].set_title('Sine Plot')
axs[0].set_ylabel('Amplitude')
axs[1].set_title('Cosine Plot')
axs[1].set_xlabel('Angle')
axs[1].set_ylabel('Amplitude')

plt.show()
```

### Vertically & Horizontally stacked sub plots.

```python
# Create some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two horizontally stacked subplots
fig, axs = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the first subplot
axs[0].plot(x, y1, 'r')

# Plot the second subplot
axs[1].plot(x, y2, 'b')

# Add titles and labels
axs[0].set_title('Sine Plot')
axs[0].set_xlabel('Angle')
axs[0].set_ylabel('Amplitude')
axs[1].set_title('Cosine Plot')
axs[1].set_xlabel('Angle')

plt.show()
```

### Stacking subplots in two directions

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2)

for i in range(2):
    for j in range(2):
        ax = axes[i, j]
        ax.plot([1, 2, 3], [4, 5, 6])
        ax.set_title(f"Subplot {i},{j}")

plt.show()
```

### Sharing of the Axes:

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        ax = axes[i, j]
        ax.plot([1, 2, 3], [4, 5, 6])
        ax.set_title(f"Subplot {i},{j}")

plt.show()
```
