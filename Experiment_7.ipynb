```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Salary_Data.csv')
```

### Univariate analysis

```python
# Univariate analysis
# Histogram of a single variable
sns.histplot(df['Age'])
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Bivariate analysis

```python
# Bivariate analysis
# Scatter plot of two variables
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Scatter plot of Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

### Multivariate analysis

```python
# Multivariate analysis
# Heatmap of correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### DBSACN Algorithm

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=30, centers=3, n_features=2, random_state=42)

# Create DBSCAN object
dbscan = DBSCAN(eps=1.5, min_samples=4)

# Fit model and predict cluster labels
labels = dbscan.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.show()
print("Number of outliers:", sum(labels==-1))
```
