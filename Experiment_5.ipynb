### Plot(x,y)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line plot')

# Show plot
plt.show()
```

### scatter(x,y)

```python
plt.scatter(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')

# Show plot
plt.show()
```

### bar(x,height)

```python
# Generate data
x = ['A', 'B', 'C', 'D']
height = [3, 6, 2, 8]

# Create plot
plt.bar(x, height)

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Vertical bar plot')

# Show plot
plt.show()
```

### stem(x,y)

```python
# Generate data
a = np.arange(0, 10, 0.1)
b = np.sin(a)

# Create plot
plt.stem(a, b)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stem plot')

# Show plot
plt.show()
```

### fill_between(x,y1,y2)

```python
p = np.arange(0, 10, 0.1)
q1 = np.sin(p)
q2 = np.cos(p)

# Create plot
plt.fill_between(p, q1, q2)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Filled area plot')

# Show plot
plt.show()
```

### stackplot(x, y)

```python
plt.stackplot(p, q1, q2)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stacked area plot')

# Show plot
plt.show()
```

### Hist(x)

```python

x = np.random.normal(size=1000)

# Create plot
plt.hist(x, bins=30)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Show plot
plt.show()
```

### boxplot(x)

```python
# Generate data
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(2, 1, 100)
data3 = np.random.normal(-2, 1, 100)

# Combine data into a list
data = [data1, data2, data3]

# Create box-and-whisker plot
plt.boxplot(data)

# Set title and labels
plt.title('Box-and-whisker plot')
plt.xlabel('Data')
plt.ylabel('Values')

# Show plot
plt.show()
```

### 4.7 To Set Axis Range (xlim, ylim) in Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# Create plot
plt.plot(x, y)

# Set axis labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine wave')

# Add annotation
plt.annotate('Maximum', xy=(np.pi/2, 1), xytext=(4, 0.8),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Add legend
plt.legend(['Sine wave'])

# Set axis range
plt.xlim([0, 10])
plt.ylim([-1.2, 1.2])

# Show plot
plt.show()
```
